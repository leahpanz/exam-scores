```python
#Identifying possible outliers
```

```python
def detect_outliers_iqr(data, WritingScore):
    Q1 = data[WritingScore].quantile(0.25)
    Q3 = data[WritingScore].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[WritingScore] < lower_bound) | (data[WritingScore] > upper_bound)]
```

```python
outliers = detect_outliers_iqr(data, 'WritingScore')
print(outliers)
```
